# 9) Convert the normal distribution to standard normal distribution for salary column

```python
import seaborn as sns 
sns.distplot(dataset["salary"])
```

```python
def stdNBgraph(dataset):

    # Calculate Mean and Standard Deviation
    mean=dataset.mean()
    std=dataset.std()

    values=[i for i in dataset] # In this oneliner, the output is converted in to list

    # Calculate the z-score using formula
    z_score = [((x-mean)/std) for x in values]

    #Plotting
    import seaborn as sns
    sns.displot(z_score,kde=True) # Kernel Density function displays the curve

    #z_score.std()
    sum(z_score)/len(z_score) 
```

```python
stdNBgraph(dataset["salary"])
```

Explanation:

- The graph displays the distribution of standardized salary values.
- The x-axis represents the standardized salary, which has been transformed to have a mean of 0 and a standard deviation of 1.
- The y-axis indicates the frequency of occurrences for each standardized salary range.

# 10) What is the probability Density Function of the salary range from 700000 to 900000?

```python
def get_pdf_probability(dataset, startrange, endrange):
    # import libraries
    import matplotlib.pyplot as plt
    from scipy.stats import norm
    import seaborn as sns

    # Distribution plot
    #axis = sns.distplot(dataset, kde=True, kde_kws={'color':'blue'}, color='Green')
    axis = sns.displot(dataset, kde=True, color='green')
    plt.axvline(startrange, color='Red')
    plt.axvline(endrange, color='Red')

    # calculate parameters - Mean and Standard Deviation
    mean =dataset.mean()
    stdDev = dataset.std()
    print("Mean: ", mean, '\nStandard Deviation: ', stdDev)
    
    # define the normal distribution of the data(inbuilt)
    Norm_dist = norm(mean, stdDev)
    
    # sample probabilities for a range of outcomes
    values = [value for value in range(startrange, endrange)]
    probabilities = [Norm_dist.pdf(value) for value in values]    
    prob = sum(probabilities)

    return prob
```

```python
get_pdf_probability(dataset["salary"],700000, 900000)
```
